# Monte Carlo Integration

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def f(x):
    y = np.sin(x)
    return y

y = f(x)
```

### Initial Conditions

```python
a = 0.0
b = np.pi

dx = 0.1

print(f'Integrating from {a} to {b}')
```

```python
N = int((b-a)/dx)

print(f'N = {N}')
```

### Defining Arrays

```python
x = np.linspace(a, b, N)
```

```python
plt.plot(x, y, 'b-')
plt.title('sin(x)');
```

### Number of Random Points

```python
B = 10000

print(f'Generating {B} points.')
```

### Random Arrays

```python
x_rand = np.zeros(B)
y_rand = np.zeros(B)

MyPointX = np.zeros(B)
MyPointY = np.zeros(B)
```

### Random Point Generator

```python
counter = 0.0

for i in range (0, B):
    x_rand[i] = np.pi * sp.rand()
    y_rand[i] = sp.rand()
    
    if y_rand[i] <= f(x_rand[i]):
        counter += 1
        MyPointX[i] = x_rand[i]
        MyPointY[i] = y_rand[i]

Int_Points = counter
print(f'The number of points below the function is {Int_Points}')
```

```python
plt.plot(x_rand, y_rand, 'r.', x, y, 'b-', MyPointX, MyPointY, 'g.');
```

```python
area = (b-a) * np.max(f(x))

print(f'The area used to integrate is {area}')
```

```python
My_Integration = (Int_Points/B) * area

print(f'The area under the function is {My_Integration}.')
```
