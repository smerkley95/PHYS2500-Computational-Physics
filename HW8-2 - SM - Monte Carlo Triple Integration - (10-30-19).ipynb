# Monte Carlo Integration of Intersecting Sphere and Cylinder

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import mpl_toolkits.mplot3d as mpl
```

### Initial Conditions

```python
r_sphere = 1.0

r_cylinder = 0.5
```

```python
t_initial = 0.0
t_final = 2*np.pi

dt = 0.1
```

```python
N = int((t_final-t_initial)/dt)

print(f'N = {N}')
```

```python
t = np.linspace(t_initial, t_final, N)
```

```python
x = r_sphere * np.cos(t)
y = r_sphere * np.sin(t)

x_cyl = r_cylinder * np.cos(t) + 0.5
y_cyl = r_cylinder * np.sin(t)
```

```python
plt.plot(x, y, 'k-',x_cyl, y_cyl, 'r-');
```

### Random Arrays

```python
x_rand = np.zeros(N)
y_rand = np.zeros(N)

MyPointX = np.zeros(N)
MyPointY = np.zeros(N)
```

### Random Point Generator

```python
counter = 0.0

for i in range(0, N):
    x_rand[i] = 2 * sp.rand() - 1
    y_rand[i] = 2 * sp.rand() - 1
    
    #if y_rand[i] >= np.min(y_cyl) and y_rand[i] <= np.max(y_cyl) and x_rand[i] >= np.min(x_cyl) and x_rand[i] <= np.max(x_cyl): 
    if x_rand[i] < 1.0 and x_rand[i] > 0.0 and y_rand[i] < 1.0 and y_rand[i] > -1.0:
        counter += 1.0
        MyPointX[i] = x_rand[i]
        MyPointY[i] = y_rand[i]
        

CylinderInt = counter
print(f'The number of points in the cylinder is {CylinderInt}')
```

```python
plt.plot(x_rand, y_rand, 'b.', x, y, 'k-', x_cyl, y_cyl, 'r-', MyPointX, MyPointY, 'g.');
plt.plot([0, 1], [1, 1], 'k-', [1, 1], [1, -1], 'k-', [1, 0], [-1, -1], 'k-', [0, 0], [-1, 1], 'k-')
```
